## Quaternion Integration
- Note that this notebook needs to be run with ros in path so no conda env is needed.
Let's try to understand the behaivour of quaternion rates and integration and compare them to euler method. 

We apply a translational and rotational velocity trajectory to a coordinate starting from the inertial frame and using both euler method and quaternions translate the frame and observer the difference and compare accuracy and more importantly pass the frame through the singular euler point and observe the difference in behaviour under both methods.  

```python
# Some imports
from numpy import *
from scipy import linalg
import matplotlib.pyplot as plt
import tf # ROS's transformation tool 
%matplotlib inline
```

```python
# Helpful functions

def s(x):
    """ Skew Operator """
    return array([[0.0, -x[2], x[1]], [x[2], 0.0, -x[0]], [-x[1], x[0], 0.0]])

def integral(x_dot, x, t):
    """ Computes the integral of x dt """
    integral = multiply(x_dot, t) + array(x).T
    # Prevent integral wind up #TODO: Check implementation, do rotational values need a different threshold?
    # integral[abs(integral) < 0.01] = 0.0
    return integral

def inverse_jacobian(p):
    """ Generate the inverse Jacobian for calculating p_dot = Je_inv*v"""
    # Generated using the awesome SymPy Je = [[bRi 0], [0 Jko]] and this is the inverse of that
    [roll_, pitch_, yaw_] = p[3:6]
    return array([[cos(pitch_)*cos(yaw_),
                   sin(pitch_)*sin(roll_)*cos(yaw_) - sin(yaw_)*cos(roll_),
                   sin(pitch_)*cos(roll_)*cos(yaw_) + sin(roll_)*sin(yaw_), 0, 0, 0],
                  [sin(yaw_)*cos(pitch_),
                   sin(pitch_)*sin(roll_)*sin(yaw_) + cos(roll_)*cos(yaw_),
                   sin(pitch_)*sin(yaw_)*cos(roll_) - sin(roll_)*cos(yaw_), 0, 0, 0],
                  [-sin(pitch_), sin(roll_)*cos(pitch_), cos(pitch_)*cos(roll_), 0, 0,
                   0],
                  [0, 0, 0, 1, sin(roll_)*tan(pitch_), cos(roll_)*tan(pitch_)],
                  [0, 0, 0, 0, cos(roll_), -sin(roll_)],
                  [0, 0, 0, 0, sin(roll_)/cos(pitch_), cos(roll_)/cos(pitch_)]], dtype=float)

def inverse_jacobian_q(p_q):
    """ Generate the inverse Jacobian using quaternions for calculating p_dot_q = Je_inv*v"""
    [qi, qj, qk, qr] = p_q
    Jeq_inv = array([[-2*qj**2 - 2*qk**2 + 1, 2*qi*qj + 2*qk*qr, 2*qi*qk - 2*qj*qr, 0,
        0, 0, 0],
       [2*qi*qj - 2*qk*qr, -2*qi**2 - 2*qk**2 + 1, 2*qi*qr + 2*qj*qk, 0,
        0, 0, 0],
       [2*qi*qk + 2*qj*qr, -2*qi*qr + 2*qj*qk, -2*qi**2 - 2*qj**2 + 1, 0,
        0, 0, 0],
       [0, 0, 0, 2.0*qr, 2.0*qk, -2.0*qj, -2.0*qi],
       [0, 0, 0, -2.0*qk, 2.0*qr, 2.0*qi, -2.0*qj],
       [0, 0, 0, 2.0*qj, -2.0*qi, 2.0*qr, -2.0*qk]],  dtype=float)
    Jeq_inv = linalg.pinv(Jeq_inv)
    return Jeq_inv

def iRb_q(p_q):
    """ Generate the SO(3) rotation matrix from body to inertial frame, using quaternion as input"""
    [qi, qj, qk, qr] = p_q

    return array([[-2*qj**2 - 2*qk**2 + 1, 2*qi*qj - 2*qk*qr, 2*qi*qk + 2*qj*qr],
       [2*qi*qj + 2*qk*qr, -2*qi**2 - 2*qk**2 + 1, -2*qi*qr + 2*qj*qk],
       [2*qi*qk - 2*qj*qr, 2*qi*qr + 2*qj*qk, -2*qi**2 - 2*qj**2 + 1]], dtype=float)

def exp_wq(v, dt):
    """ Compute the exponential e^(Adt/2)=Exp{w dt}=q{w dt}"""
    # Norm of v
    vn = linalg.norm(v)
    # This can be singular, but we know when.
    if vn < 0.0001:
        return array([0,0,0,1])
    else:
        cs = cos(vn*dt/2)
        sn_n = sin(vn*dt/2)/vn
        return array([v[0]*sn_n, v[1]*sn_n, v[2]*sn_n, cs])
```

```python
# Setup the trajectory - Let's have the vehicle initially surge forward and do some mixed rotations

Ts = .01
Tfinal = 5
samples = arange(0,int(Tfinal/Ts))
v = zeros((6, len(samples)), dtype=float64)

for k in samples:
    v[0,k] = 0 # Surge velocity (m/s)
    v[3,k] = 0 # Roll Rate (rad/s)
    v[4,k] = 10*sin(4*2*pi*k/len(samples)) # Pitch Rate (rad/s)
    v[5,k] = 0#5*sin(2*pi*k/len(samples)) # Yaw Rate (rad/s)
    # Constants for now, but change later
```

```python
# First we try the euler method
q0 = array([0,0,0,0,0,0]) 
q = zeros((6, len(samples)), dtype=float)
q[:,0] = q0
for k in arange(1, len(samples)):
    Je_inv = inverse_jacobian(q[:,k-1])
    q_dot = dot(Je_inv, v[:,k-1])
    q[:,k] = integral(q_dot, q[:,k-1], Ts)

fig = plt.figure(figsize=(20,10))
labels = ['surge', 'sway', 'heave', 'roll', 'pitch', 'yaw']
labels_r = ['surge rate', 'roll rate', 'pitch rate', 'yaw rate']
# Add singularity lines
plt.subplot(221)
plt.ylim([-2,2])
plt.subplot(222)  
plt.plot(samples, ones(len(samples))*pi/2)
plt.plot(samples, -ones(len(samples))*pi/2)

# Add bounds lines
plt.subplot(222)  
plt.plot(samples, ones(len(samples))*pi)
plt.plot(samples, -ones(len(samples))*pi)
for k in arange(0,6):
    lw_ = 1
    # highlight pitch line
    if k == 4:
        lw_ = 2
    if k < 3:
        plt.subplot(221)
    else:
        plt.subplot(222)    
    plt.plot(q[k,:], lw = lw_, label = labels[k])
    plt.legend()
plt.subplot(223)
plt.ylim([-10,10])
for k in arange(0,6):
    if k == 0:
        plt.plot(v[k,:], lw =2, label = labels_r[k])
    if k > 2:
        plt.plot(v[k,:], lw =2, label = labels_r[k-2])
    plt.legend()
plt.show()
# p_q = tf.transformations.quaternion_from_euler(self.p[3], self.p[4], self.p[5], 'rxyz')
```

As you can see, when pitch reachs -pi/2, pi/2 it freaks out and the other positions jump! 

```python
# Let's do it the quaternion way
q0_q = zeros(7) # Zero initial position
q_q = zeros((7, len(samples)))
q_q[:,0] = q0_q.copy()
q_q_euler = zeros((6, len(samples)))
q_q_euler[:,0] = q0
q_q[3:7,0] = tf.transformations.quaternion_from_euler(q0[3],q0[4],q0[5], 'rxyz')
print(q_q[:,0])
expq = zeros((4, len(samples)))
for k in arange(1, len(samples)):
    # We take the p tran rotation from the inverse jacobian
    Rinv = iRb_q(q_q[3:7,k-1])
    q_dot_q = dot(Rinv, v[0:3, k-1])
    q_q[0:3,k] = integral(q_dot_q, q_q[0:3,k-1], Ts)
    expq[:,k] = exp_wq(v[3:6, k-1], Ts)
    q_q[3:7,k] = tf.transformations.quaternion_multiply(q_q[3:7:,k-1], expq[:,k])
    # check sign and invert
    qconj = tf.transformations.quaternion_conjugate(q_q[3:7,k])
    
    qd = tf.transformations.quaternion_multiply(q_q[3:7, k-1], qconj)
#     if qd[3] < 0:
#         q_q[3:7,k] = -q_q[3:7, k]
    q_q_euler[0:3,k] = q_q[0:3, k]
    q_q_euler[3:6,k] = tf.transformations.euler_from_quaternion(q_q[3:7, k], 'rxyz')
#     if q_q_euler[5,k] < 0:
#         print("Hello, k=", k)

    
# Now we plot, which ironically takes more lines !
fig = plt.figure(figsize=(20,10))
labels = ['surge', 'sway', 'heave', 'roll', 'pitch', 'yaw']
labels_r = ['surge rate', 'roll rate', 'pitch rate', 'yaw rate']
# Add singularity lines
plt.subplot(122)  
plt.plot(samples, ones(len(samples))*pi/2)
plt.plot(samples, -ones(len(samples))*pi/2)

# Add bounds lines
plt.subplot(122)  
plt.plot(samples, ones(len(samples))*pi)
plt.plot(samples, -ones(len(samples))*pi)

for k in arange(0,6):
    lw_ = 1
    # highlight pitch line
    if k == 4:
        lw_ = 2
    if k < 3:
        plt.subplot(221)
    else:
        plt.subplot(122)    
    plt.plot(q_q_euler[k,:], lw = lw_, label = labels[k])
    plt.legend()
plt.subplot(223)
plt.ylim([-10,10])
for k in arange(0,6):
    if k == 0:
        plt.plot(v[k,:], lw =2, label = labels_r[k])
    if k > 2:
        plt.plot(v[k,:], lw =2, label = labels_r[k-2])
    plt.legend()
```

The culprit seems to be the transformation function from quaternion to euler, since the mapping is not unique I need to make sure the rotation is done in the direction of the previous att. TODO

Note that this may only be an issue if I have one of the att angles near [-pi/2 | pi/2], this is where I don't get a unique solution sometimes (reference)[http://citeseerx.ist.psu.edu/viewdoc/download?doi=10.1.1.371.6578&rep=rep1&type=pdf]. From the reference, there are two off cases, when sin(pitch) = +/-1 (when cos(pitch) = 0) there is an infinite number of solutions. Otherwise there are two solutions. So the issue does not stem from quaternions, but rather, from the rotation matrix back to euler conversion. Which again is why I should use quaternions to control my vehicle and only use euler to command it since there is no issue getting a rotation matrix from euler angles.
